```python
import matplotlib.pyplot as plt
    
def CalcuateRows(TotalPlots):
    if TotalPlots==3:
        return 1
    elif TotalPlots%3 == 0:
        return(int(TotalPlots/3))
    else:
        remainder = TotalPlots%3
        return(TotalPlots//3 + remainder)

# DistPlot 

TotalDistPlots = len(con_features)
f, axes = plt.subplots(CalcuateRows(TotalDistPlots), 3, figsize=(10,10))

column = 0
row = 0
while(row<TotalDistPlots):
    for i in con_features:
        if column<3:
            axes[row, column].set_title('DistPlot '+ i)
            sns.distplot(tips[i],ax=axes[row,column])	
            column+=1
        if column==3:
            row+=1
            column=0
    row+=1
plt.savefig("DistPlots.pdf")
plt.tight_layout()
```

```python
cat_features = ['Street']
con_features=['SalePrice','LotFrontage']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
    
def CalcuateRows(TotalPlots):
    if TotalPlots<=3:
        return 2
    elif TotalPlots%3 == 0:
        return(int(TotalPlots/3))
    else:
        return(TotalPlots//3 + 1)


TotalBarPlots = len(cat_features)
f, axes = plt.subplots(CalcuateRows(TotalBarPlots), 3, figsize=(10,10))

column = 0
row = 0

while(row<TotalBarPlots):
    for i in con_features:
        for j in cat_features:
                if column<3:
                    try:
                        axes[row, column].set_title('BarPlot '+ i + ' '+j)
                        sns.barplot(x=i,y=j,data=df,ax=axes[row,column])
                        column+=1
                        if column==3:
                            row+=1
                            column=0
                    except:
                        print("",end="")
    row+=1
plt.savefig("BarPlots.pdf")
plt.tight_layout()
```
