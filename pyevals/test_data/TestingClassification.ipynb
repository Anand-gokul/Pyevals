```python
import pyevals 
```

```python
import pandas as pd

data = pd.read_csv("titanic.csv")
```

```python
data.isnull().any()
```

```python
data = data.drop(["Age", "Cabin", "Embarked", "Name", "Sex", "Ticket"], 1)
```

```python
X = data.drop("Survived", 1)
y = data["Survived"]
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.2)
```

```python
a = pyevals.build(X_train, X_test, y_train, y_test, 'classification')
```

```python
a.evaluate()
```
